```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df= pd.read_csv('/content/train_and_test2.csv')
print(df.head())
sns.countplot(x='Sex' , data=df)
plt.title("Count Plot of Sex")
plt.show()
```

```python
df['Sex'].value_counts().plot.pie(autopct= '%1.1f%%')
plt.ylabel('')
plt.title("Gender distribution")
plt.show()
```

```python
plt.hist(df['Age'].dropna() ,bins =20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.histplot(df['Age'].dropna(),kde= True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(y='Sex' ,data = df)
plt.show()
```
